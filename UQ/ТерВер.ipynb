## Непрерывная случайная величина
Непрерывная случайная величина задается плотностью распределения

**Пример непрерывной случайной величины**

Норма́льное распределе́ние[1][2], также называемое распределением Гаусса или Гаусса — Лапласа[3] — распределение вероятностей, которое в одномерном случае задаётся **функцией плотности вероятности**, совпадающей с функцией Гаусса:
$f(x)=\frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x-\mu)^{2}}{2 \sigma^{2}}}$

**Стандартным нормальным** распределением называется нормальное распределение с математическим ожиданием μ = 0 и стандартным отклонением σ = 1.

# Ортогональные полиномы
## Скалярное произведение в n-мерном пространстве
$\overline{a}=\left(a^{1}, a^{2}, \ldots, a^{n}\right)$ - n-мерный вектор

$\overline{b}=\left(b^{1}, b^{2}, \ldots, b^{n}\right)$

$\overline{a} \cdot \overline{b}=a^{1} b^{1}+a^{2} b^{2}+\ldots+a^{n} b^{n}$

Говоря об бесконечно мерном простарнстве, $a, b - $ ни что иное как кортеж или последовательность
$\overline{a}=\left(a^{1}, a^{2}, \ldots, a^{n}, \ldots\right)=\left\{a_{n}\right\}= \left\{a_{n}\right\}_{n \in \mathbb{N}}$ (1)

$\overline{b}=\left(b^{1}, b^{2}, \ldots, b^{n}, \ldots\right)=\left\{b_{n}\right\} = \left\{b_{n}\right\}_{n \in \mathbb{N}}$ (2), где $\mathbb{N}$ - бесконечное множество натуральных чисел

Очевидно, скалярное произведение будет иметь следующий вид:
$\overline{a} \cdot \overline{b}=a^{1} b^{1}+a^{2} b^{2}+\ldots+a^{n} b^{n}+ \ldots=\sum_{i=1}^{\infty} a_{i} b_{i}$

Перейдем от счетного множества к континуальному множесту. **Конти́нуум** в теории множеств — мощность (размерность) множества всех вещественных чисел. Тогда (1) - $\overline{a}=\left\{a_{x}\right\}_{x \in \mathbb{K}}$  (2) - $\overline{b}=\left\{b_{x}\right\}_{x \in \mathbb{K}}$  (здесь индексное множество пробегает континум)

Теперь мы понимаем, что вместо того чтобы говорить о вот таких последовательностях, удобно говорить об $a$, зависящем от $x$

$a(x)=\left\{a_{x}\right\}_{x \in \mathbb{K}}$ - линейное пространство

$b(x)=\left\{b_{x}\right\}_{x \in \mathbb{K}}$ - линейное пространство

Тогда, когда мы перешли от счетных наборов (дискретного множетсва) к континуальным наборам - сумма переходит на интеграл

$(a, b)=\int_{K} a(x) b(x)$ (дифференциал в конце не указан)

**Скалярные произведения с весом**

$(a, b)=\int_{K} \rho(x) a(x) b(x)$, где $\rho(x)$ - это вес, зависящий от значения $x$


## Ортогонализация полиномов
Рассмотрим список мономов:
$1, x, x^2, x^3, x^4$. Рассмотрим отрезок $K = [-1,+1]$. Очевидно, что полиномы друг с другом не ортогональны на данном симметричном отрезке (в заданом списке мономов сущестуют четные функции, а интеграл от четных функций равен 0). Процесс ортогонализации неортогональных полиномов называется процесс **Грамма-Шмидта**. Процесс рассматривает следа (проекции) от предыдущих векторов и рассчитывает новый вектор, оротогональный предыдущим векторам (выходя за границы 3-х мерного пространства). Реккурентная формула для расчета ортогонального вектора выглядит следующим образом:

$P_{n+1}(x)=\frac{2 n+1}{n+1} x P_{n}(x)-\frac{n}{n+1} P_{n-1}(x)$


## Ковариация

Если между случайными величинами $\xi$ и $\eta$ существует стохастическая связь, то одним из парамтеров, характеризующим меру этой связи является ковариация cov$(\xi, \eta)$. Формула следующая

$\operatorname{cov}(\xi, \eta)=M[(\xi-M \xi) \cdot(\eta-M \eta)] \equiv M(\xi \eta)-M \xi \cdot M \eta$

Если данные величины независимы, то $\operatorname{cov}(\xi, \eta) =0$ **Обратное неверно!!!** Из равенства нулю ковариации **не следует независимость случайных величин**. Случайные величины могут быть зависимыми в то время как их ковариация — нулевая! 

Но зато, если ковариация случайных величин отлична от нуля, то между ними существует **стохастическая связь**, мерой которой и является величина ковариации.

$\operatorname{cov}(\xi, \xi)=D \xi$ и $\operatorname{cov}(\eta, \eta)=D \eta$

## Матрица ковариаций

Ковариационной матрицей случайного вектора $(\xi, \eta)$\ (случайный вектор состоит из случайных величин) -  имеет следующий вид:

$\operatorname{Cov}(\xi, \eta)=\left[\begin{array}{ll}{\operatorname{cov}(\xi, \xi)} & {\operatorname{cov}(\xi, \eta)} \\ {\operatorname{cov}(\eta, \xi)} & {\operatorname{cov}(\eta, \eta)}\end{array}\right] \equiv\left[\begin{array}{cc}{\sigma_{\xi}^{2}} & {\operatorname{cov}(\xi, \eta)} \\ {\operatorname{cov}(\eta, \xi)} & {\sigma_{\eta}^{2}}\end{array}\right]$
